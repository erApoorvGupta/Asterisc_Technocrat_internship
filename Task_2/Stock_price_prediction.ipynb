# <span style="color: purple; font-size: 56px; font-weight: bold;">LSTM Networks for Stock Prediction</span>

</div>

![LSTM Network](https://upload.wikimedia.org/wikipedia/commons/thumb/9/93/LSTM_Cell.svg/1200px-LSTM_Cell.svg.png)


#### LSTM networks have gained popularity in stock prediction tasks due to their ability to capture long-term dependencies and handle sequential data effectively. Stock prices are inherently sequential, where each data point depends on previous data points and exhibits temporal patterns. Traditional machine learning models struggle to capture these complex patterns, leading to suboptimal predictions.

#### LSTM networks excel at modeling and predicting time series data, making them a natural choice for stock prediction. Their memory cell allows them to remember important historical information, such as past stock prices, trends, and market conditions. By learning from this historical context, LSTM networks can identify and exploit patterns in the data that may affect future stock prices.

#### Furthermore, LSTM networks can handle noisy and non-linear data, which are common characteristics of stock market data. They can adapt and update their internal state based on new information, enabling them to adapt to changing market dynamics. This adaptability is crucial for accurate stock price predictions, as market conditions and trends evolve over time.

#### Another advantage of LSTM networks is their ability to incorporate various input features. In addition to historical stock prices, other relevant factors such as trading volume, technical indicators, news sentiment, and economic data can be included as input to the network. This flexibility allows LSTM networks to capture multiple dimensions of information and potentially improve the prediction accuracy.



```python
# Applying LSTM model for stock price prediction
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Extract the 'Adj Close' column as the target variable
target_variable = 'Adj Close'
target_data = df[target_variable].values

# Split the data into training and testing sets (80% train, 20% test)
train_size = int(len(df) * 0.8)
train_data = target_data[:train_size]
test_data = target_data[train_size:]

# Normalize the data to a range between 0 and 1
scaler = MinMaxScaler(feature_range=(0, 1))
train_data_normalized = scaler.fit_transform(train_data.reshape(-1, 1))
test_data_normalized = scaler.transform(test_data.reshape(-1, 1))

# Reshape the data into the input shape required by LSTM models (samples, time steps, features)
def create_sequences(data, time_steps):
    X = []
    y = []
    for i in range(len(data) - time_steps):
        X.append(data[i : i + time_steps])
        y.append(data[i + time_steps])
    return np.array(X), np.array(y)

time_steps = 2  # Adjust this value as per your requirement
X_train, y_train = create_sequences(train_data_normalized, time_steps)
X_test, y_test = create_sequences(test_data_normalized, time_steps)

# Build the LSTM model architecture
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(time_steps, 1), return_sequences=True))
model.add(LSTM(units=50, activation='relu', return_sequences=True))
model.add(LSTM(units=50, activation='relu', return_sequences=True))
model.add(LSTM(units=50, activation='relu'))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model on the training data
model.fit(X_train, y_train, epochs=500, batch_size=32)

# Evaluate the model on the testing data
loss = model.evaluate(X_test, y_test)
print(f'Test Loss: {loss}')

# Make predictions using the trained model
predicted_data_normalized = model.predict(X_test)
predicted_data = scaler.inverse_transform(predicted_data_normalized)

```

```python
model.summary()
```

### Evaluation Metrics

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calculate the predictions
predicted_data_normalized = model.predict(X_test)
predicted_data = scaler.inverse_transform(predicted_data_normalized)

# Calculate the evaluation metrics
mse = mean_squared_error(y_test, predicted_data)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, predicted_data)
r2 = r2_score(y_test, predicted_data)

# Print the evaluation metrics
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'R-squared (R^2) Score: {r2}')
```

```python
# Get the predicted values for the training dataset
predicted_train_normalized = model.predict(X_train)
predicted_train = scaler.inverse_transform(predicted_train_normalized)

# Get the predicted values for the test dataset
predicted_test_normalized = model.predict(X_test)
predicted_test = scaler.inverse_transform(predicted_test_normalized)

# Plot the predicted and real values for the training dataset
plt.figure(figsize=(15, 5))
plt.plot(predicted_train, label='Predicted (Train)', color='blue')
plt.plot(target_data[:len(X_train)], label='Real (Train)', color='green')

# Plot the predicted and real values for the test dataset
test_range = range(len(X_train), len(X_train) + len(predicted_test))
plt.plot(test_range[:len(predicted_test)], predicted_test, label='Predicted (Test)', color='red')
plt.plot(test_range[:len(predicted_test)], target_data[-len(predicted_test):], label='Real (Test)', color='orange')

plt.xlabel('Time', fontsize=14, fontweight='bold')
plt.ylabel('Adj Close', fontsize=14, fontweight='bold')
plt.title('Predicted vs Real Stock Prices', fontsize=24, fontweight='bold',color='purple')
plt.legend()

# Add grid lines
plt.grid(True, which='both', linestyle='-', linewidth=0.5)

# Customize tick labels
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(8))
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(5))

legend = plt.legend(frameon=True, shadow=True)

for line in legend.get_lines():
    line.set_linewidth(1.5)

plt.tight_layout()
plt.show()
```

#### <b> The graph compares the predicted and real stock prices over time. The blue line represents the predicted prices for the training dataset, while the green line represents the actual prices for the training dataset. The red line shows the predicted prices for the test dataset, plotted against the actual prices (shown in orange) for the test dataset. The x-axis represents time, and the y-axis represents the adjusted closing prices of the stocks. This graph provides an overview of how well the LSTM model captures the underlying patterns and trends in the data.
