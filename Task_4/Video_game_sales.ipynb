```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

### Information about the Dataset



```python
# loading Dataset
df = pd.read_csv('Video_Games_Sales_task_3.csv')
df
```

```python
df.info()
```

<br>

### There is missing data.. The data needs cleaning

```python
df.isnull().sum()
```

```python
# finding percentage of null values
df.isnull().sum()/df.shape[0]*100
```

### Dropping 'Name' column as it is not required

```python
# dropping name column
df.drop('Name',axis=1,inplace=True)
```

### Testing correlation and understanding the data

```python
df_na = df.dropna()
df_na.info()
```

### We label encode the data to see possible relations with 'Global_Sales'

```python
# label encode df_na
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_na['Genre'] = le.fit_transform(df_na['Genre'])
df_na['Publisher'] = le.fit_transform(df_na['Publisher'])
df_na['Platform'] = le.fit_transform(df_na['Platform'])
df_na['Rating'] = le.fit_transform(df_na['Rating'])
df_na['Developer'] = le.fit_transform(df_na['Developer'])
```

<br>

### We try to compare correlation form the complete data in order to get a better understanding of the data and then we try to drop or impute coloumn depending its impact on the column 'Global_Sales'

```python
# correlation matrix for df_na in heatmap    
plt.figure(figsize=(10,10))
sns.heatmap(df_na.corr(), annot=True, cmap='coolwarm')
```

### After seeing the correlations and considering the number of null values in each column. We can colude to drop 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Developer' and 'Rating'

1. More than 40% of the data is missing in 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Developer' and 'Rating' columns. So we dont have any concrete beliefs that this data will help us in predicting the 'Global_Sales' column.

2. The correlation of 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Developer' and 'Rating' columns with 'Global_Sales' is also very low.

```python
# dropping columns 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Developer' and 'Rating'
df = df.drop(['Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Developer', 'Rating'], axis=1)
```

<br><br>

### As the number of rows with null values is very low, we can now efficiently clean and drop them.

```python
# dropping null values
df = df.dropna()
```

### The Data is now ready for prediction

```python
df.info()
```

<br><br>

### Multiple Regression Model

```python
# Applying multiple linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# label encoding the data

# Encoding the categorical columns
df['Genre'] = le.fit_transform(df['Genre'])
df['Publisher'] = le.fit_transform(df['Publisher'])
df['Platform'] = le.fit_transform(df['Platform'])

# Splitting the data into training and testing data
X = df.drop('Global_Sales', axis=1)
y = df['Global_Sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Creating the model
model = LinearRegression()

# Fitting the model
model.fit(X_train, y_train)

# Predicting the model
y_pred = model.predict(X_test)
```

```python
# Finding the accuracy of the model
MLR_accuracy = model.score(X_test, y_test)
MLR_accuracy
```

```python
import matplotlib.pyplot as plt

# Creating a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 3))

axes[0].scatter(range(len(y_test)), y_test, color='blue', edgecolor='black', linewidths=0.2, label='Actual', s=5)
axes[0].set_xlabel('Sample Index', fontweight='bold')
axes[0].set_ylabel('Global Sales', fontweight='bold')
axes[0].set_title('Actual Data', fontsize=20, fontweight='bold', color='purple')

axes[1].scatter(range(len(y_test)), y_pred, color='red', edgecolor='black', linewidths=0.2, label='Predicted', s=5)
axes[1].set_xlabel('Sample Index', fontweight='bold')
axes[1].set_ylabel('Global Sales', fontweight='bold')
axes[1].set_title('Predicted Data', fontsize=20, fontweight='bold', color='purple')

plt.suptitle('MLR Results', fontsize=28, fontweight='bold', color='blue')

plt.tight_layout()
plt.show()
```

### <b> The final graph compares the actual values with the predicted values for the 'Global_Sales' column using Multiple Regression Model and we find that the model is quite accurate. Now there is a possibility that the model is overfitting the data. So we will use other models to check the accuracy of the model. </b>

<br><br>

### XGBoost Model

```python
# xg boost regressor

# Importing the XGBRegressor class
from xgboost import XGBRegressor

# Creating the model
model = XGBRegressor()

# Fitting the model
model.fit(X_train, y_train)

# Predicting the model
y_pred = model.predict(X_test)

# Creating a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 3))

axes[0].scatter(range(len(y_test)), y_test, color='blue', edgecolor='black', linewidths=0.2, label='Actual', s=5)
axes[0].set_xlabel('Sample Index', fontweight='bold')
axes[0].set_ylabel('Global Sales', fontweight='bold')
axes[0].set_title('Actual Data', fontsize=20, fontweight='bold', color='purple')

axes[1].scatter(range(len(y_test)), y_pred, color='red', edgecolor='black', linewidths=0.2, label='Predicted', s=5)
axes[1].set_xlabel('Sample Index', fontweight='bold')
axes[1].set_ylabel('Global Sales', fontweight='bold')
axes[1].set_title('Predicted Data', fontsize=20, fontweight='bold', color='purple')

plt.suptitle('XGboost Results', fontsize=28, fontweight='bold', color='blue')

plt.tight_layout()
plt.show()
```

```python
# finding the accuracy of the model
from sklearn.metrics import r2_score
XG_accuracy = r2_score(y_test, y_pred)
XG_accuracy
```

<br><br>

### Random Forest Model

```python
# Random Forest Regressor

# Importing the RandomForestRegressor class
from sklearn.ensemble import RandomForestRegressor

# Creating the model
model = RandomForestRegressor()

# Fitting the model
model.fit(X_train, y_train)

# Predicting the model
y_pred = model.predict(X_test)

# Creating a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 3))


axes[0].scatter(range(len(y_test)), y_test, color='blue', edgecolor='black', linewidths=0.2, label='Actual', s=5)
axes[0].set_xlabel('Sample Index', fontweight='bold')
axes[0].set_ylabel('Global Sales', fontweight='bold')
axes[0].set_title('Actual Data', fontsize=20, fontweight='bold', color='purple')


axes[1].scatter(range(len(y_test)), y_pred, color='red', edgecolor='black', linewidths=0.2, label='Predicted', s=5)
axes[1].set_xlabel('Sample Index', fontweight='bold')
axes[1].set_ylabel('Global Sales', fontweight='bold')
axes[1].set_title('Predicted Data', fontsize=20, fontweight='bold', color='purple')

plt.suptitle('Random Forest Results', fontsize=28, fontweight='bold', color='blue')

plt.tight_layout()
plt.show()
```

```python
# finding the accuracy of the model
from sklearn.metrics import r2_score
RF_accuracy = r2_score(y_test, y_pred)
RF_accuracy
```

## Comapring results of all the models

```python
# Bar plot for accuracy of all the models

# Data for the models and accuracy
models = ['MLR', 'XGBoost', 'Random Forest']
accuracy = [MLR_accuracy, XG_accuracy, RF_accuracy]
colors = ['red', 'green', 'blue']

# Creating a figure with a bar plot
plt.figure(figsize=(10, 5))
bars = plt.bar(models, accuracy, color=colors)

# Adding data labels to the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

# Customizing the plot
plt.xlabel('Models', fontsize=12, fontweight='bold')
plt.ylabel('Accuracy', fontsize=12, fontweight='bold')
plt.title('Accuracy of all the models', fontsize=16, fontweight='bold', color='purple')

# Setting the y-axis limit to highlight differences
plt.ylim([0, 1])

# Adding grid lines
plt.grid(True, axis='y', linestyle='--', linewidth=0.5)


# Adding a background color
plt.gca().set_facecolor('lightgray')

# Removing spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Displaying the plot
plt.show()
```
